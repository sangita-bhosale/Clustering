```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
airlines_data=pd.read_csv('EastWestAirlines.csv')
```

```python
airline_data2=airlines_data.drop(['ID#'],axis=1)
airline_data2
```

## # Normalize heterogenous numerical data 


```python
airline_data2_norm=pd.DataFrame(normalize(airline_data2),columns=airline_data2.columns)
airline_data2_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(airline_data2_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(airline_data2_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
airline_data2['clustersid']=hclusters.labels_
airline_data2
```

```python
airline_data2.groupby('clustersid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(airline_data2['clustersid'],airline_data2['Balance'], c=hclusters.labels_) 
```
